# Baseline DSPy generation

Notebook minimal pour inspecter les graines source, lancer le script baseline et verifier les sorties generees.


## 1. Inspecter les donnees de depart et le schema cible


```python
from pathlib import Path
import json

ROOT = Path.cwd().resolve().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd() / 'dspy_gepa'
source_path = ROOT / 'datasets' / 'source' / 'manual_seed_examples.jsonl'
schema_path = ROOT / 'schemas' / 'synthetic_record.schema.json'

print(source_path.read_text(encoding='utf-8'))
print(json.dumps(json.loads(schema_path.read_text(encoding='utf-8')), indent=2, ensure_ascii=False))
```

## 2. Lancer la baseline DSPy sans GEPA

Definir au prealable les variables d'environnement necessaires a DSPy (`DSPY_MODEL`, `OPENAI_API_KEY` et eventuellement `OPENAI_BASE_URL`).


```python
import os
import subprocess

command = [
    'python3',
    str(ROOT / 'scripts' / 'generate_baseline.py'),
    '--source', str(source_path),
    '--output', str(ROOT / 'datasets' / 'generated' / 'baseline_generations.jsonl'),
    '--limit', '2',
]
subprocess.run(command, check=False, env=os.environ.copy())
```

## 3. Valider les sorties generees


```python
generated_path = ROOT / 'datasets' / 'generated' / 'baseline_generations.jsonl'
command = ['python3', str(ROOT / 'scripts' / 'validate_generated.py'), str(generated_path)]
subprocess.run(command, check=False)
```
