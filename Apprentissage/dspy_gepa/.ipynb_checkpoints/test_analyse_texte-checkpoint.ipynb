```python
import re
from typing import Any, Dict, List, Optional
import uuid
```

```python
HASHTAG_RE = re.compile(r"(?<!\w)#([^\s#]+)")
DATE_RE = re.compile(r"\b\d{2}-\d{2}-\d{4}\b")
URL_RE = re.compile(r"\bhttps?://[^\s<>\"]+")
SPACE_RE = re.compile(r"[ \t]+")
MULTI_NL_RE = re.compile(r"\n{3,}")


def _normalize_text(text: str) -> str:
    text = text.replace("\r\n", "\n").replace("\r", "\n")
    text = SPACE_RE.sub(" ", text)
    text = MULTI_NL_RE.sub("\n\n", text)
    return text.strip()


def _remove_spans(text: str, spans: List[tuple[int, int]]) -> str:
    if not spans:
        return text

    spans = sorted(spans)
    merged = []
    cur_start, cur_end = spans[0]

    for start, end in spans[1:]:
        if start <= cur_end:
            cur_end = max(cur_end, end)
        else:
            merged.append((cur_start, cur_end))
            cur_start, cur_end = start, end
    merged.append((cur_start, cur_end))

    parts = []
    last = 0
    for start, end in merged:
        parts.append(text[last:start])
        last = end
    parts.append(text[last:])

    return "".join(parts)


def _compute_incipit_title(body: str, max_len: int = 40) -> Optional[str]:
    lines = [line.strip() for line in body.splitlines() if line.strip()]
    if not lines:
        return None

    first = lines[0]
    if len(first) <= max_len:
        return first

    cut = first[:max_len]
    if " " in cut:
        cut = cut.rsplit(" ", 1)[0]
    return cut.rstrip(" ,;:-") + "…"


def _detect_explicit_title(lines: List[str]) -> tuple[Optional[str], Optional[int]]:
    """
    Heuristique :
    - une ligne non vide
    - pas hashtag pur, pas date pure, pas URL pure
    - assez courte
    - suivie d'une ligne plus longue ou d'un paragraphe
    """
    for i in range(len(lines) - 1):
        cur = lines[i].strip()
        nxt = lines[i + 1].strip()

        if not cur:
            continue
        if not nxt:
            continue
        if HASHTAG_RE.fullmatch(cur):
            continue
        if DATE_RE.fullmatch(cur):
            continue
        if URL_RE.fullmatch(cur):
            continue

        if (
            len(cur) <= 120
            and not cur.endswith((".", "!", "?", ";"))
            and len(nxt) > len(cur)
        ):
            return cur, i

    return None, None


def analyse_texte_documentaire(text: str) -> Dict[str, Any]:
    """
    Analyse un texte libre et retourne une structure JSON-compatible :
    {
      "mots_clefs": [...],
      "titre": "...",
      "date": "...",
      "url": "...",
      "texte": "corps expurgé"
    }
    """

    original = text or ""
    text = _normalize_text(original)

    # 1. Extraire hashtags, date, url avec positions
    mots_clefs = []
    spans_to_remove: List[tuple[int, int]] = []

    for m in HASHTAG_RE.finditer(text):
        mots_clefs.append("#" + m.group(1))
        spans_to_remove.append((m.start(), m.end()))

    date_match = DATE_RE.search(text)
    date = date_match.group(0) if date_match else None
    if date_match:
        spans_to_remove.append((date_match.start(), date_match.end()))

    url_match = URL_RE.search(text)
    url = url_match.group(0) if url_match else None
    if url_match:
        spans_to_remove.append((url_match.start(), url_match.end()))

    # Dédupliquer les mots-clefs en conservant l'ordre
    seen = set()
    mots_clefs_uniques = []
    for tag in mots_clefs:
        if tag not in seen:
            seen.add(tag)
            mots_clefs_uniques.append(tag)

    # 2. Supprimer ces métadonnées du texte
    cleaned = _remove_spans(text, spans_to_remove)

    # Nettoyage après suppression
    cleaned = re.sub(r"[ \t]+\n", "\n", cleaned)
    cleaned = re.sub(r"\n[ \t]+", "\n", cleaned)
    cleaned = re.sub(r" {2,}", " ", cleaned)
    cleaned = re.sub(r"\n{3,}", "\n\n", cleaned)
    cleaned = cleaned.strip(" \n\t-:")

    # 3. Détecter un titre explicite éventuel
    lines = [line for line in cleaned.splitlines()]
    titre, title_idx = _detect_explicit_title(lines)

    if titre is not None and title_idx is not None:
        body_lines = [line for i, line in enumerate(lines) if i != title_idx]
        body = "\n".join(body_lines).strip()
    else:
        body = cleaned.strip()
        titre = _compute_incipit_title(body)

    body = _normalize_text(body)

    return {
        "id": str(uuid.uuid4()),
        "mots_clefs": mots_clefs_uniques,
        "titre": titre,
        "date": date,
        "url": url,
        "texte": body,
    }
```

```python
texte = """
#histoire #France
La Révolution française
14-07-1789
https://archives.fr/revolution

La Révolution française marque une rupture politique et sociale majeure.
Elle met fin à l'Ancien Régime et transforme durablement les institutions.
"""

resultat = analyse_texte_documentaire(texte)
print(resultat)
```

```python
import ast
import json
from pathlib import Path

input_path = Path("corpus-100.jsonl")
output_path = Path("corpus-100-traité.jsonl")

nb_lignes = 0
nb_erreurs = 0

with input_path.open("r", encoding="utf-8") as fin, output_path.open("w", encoding="utf-8") as fout:
    for no_ligne, ligne in enumerate(fin, start=1):
        ligne = ligne.strip()
        if not ligne:
            continue

        try:
            enregistrement = ast.literal_eval(ligne)

            if not isinstance(enregistrement, dict):
                raise ValueError("la ligne ne contient pas un dictionnaire")

            if "texte" not in enregistrement:
                raise ValueError("champ 'texte' absent")

            texte = enregistrement["texte"]
            resultat = analyse_texte_documentaire(texte)

            if nb_lignes % 10 == 0:
                print(resultat)
            fout.write(json.dumps(resultat, ensure_ascii=False) + "\n")
            nb_lignes += 1

        except Exception as e:
            nb_erreurs += 1
            print(f"Erreur ligne {no_ligne}: {e}")

print(f"Terminé. {nb_lignes} lignes traitées, {nb_erreurs} erreurs.")
print(f"Fichier de sortie : {output_path}")
```
